## Reading and Preparing the Data

```python
df=pd.read_csv('data_capstone_2/nlp_reviews_cleaned.csv', delimiter=',')
```

#### Spliting the ratings into two class

```python
df['rating_class'] = df['rating'].apply(lambda x: 0 if x <= 2 else 1)
print(df.rating_class.value_counts())
```

## Getting initial Score

### Creating base models with CountVectorizer (unigram)

```python
df_train = df[0:10000]
```

```python
X_1 = df_train['clean_text']
y_1 = df_train['rating_class']
X_train1, X_test1, y_train1, y_test1 = train_test_split(X_1, y_1, test_size=0.25, random_state=5)

cv1 = CountVectorizer(ngram_range=(1,1))
cv_train1 = cv1.fit_transform(X_train1)
cv_train1 = cv_train1.toarray()
cv_test1 = cv1.transform(X_test1)
cv_test1 = cv_test1.toarray()
```

```python
from sklearn import model_selection
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC

clf1 = LogisticRegression(random_state=1)
clf2 = SVC(kernel = 'linear')
clf3 = RandomForestClassifier(random_state=1)


labels = ['Logistic Regression', 'Linear SVM', 'Random Forest']
print('Initial Scores\n\n')

for clf, label in zip([clf1, clf2, clf3], labels):
    clf.fit(cv_train1, y_train1)
    y_pred_clf = clf.predict(cv_test1)
    cm = confusion_matrix(y_test1, y_pred_clf)
    
    print('********** [{}] **********\n'.format(label))
    print('1. Accuarcy: {}\n'.format(metrics.accuracy_score(y_test1, y_pred_clf)))
    print('2. The F-1 score of the model {}\n'.format(f1_score(y_test1, y_pred_clf, average='weighted')))
    print('3. The recall score of the model {}\n'.format(recall_score(y_test1, y_pred_clf, average='weighted')))
    print('4. Classification Report:\n{}\n5. Confusion matrix:\n{}\n\n\n'.format
          (classification_report(y_test1, y_pred_clf), cm))
```

#### Logistic Regression has the best f-1 score of model: 0.89 
#### Linear SVM has the best f-1 score of minorty class: 0.39
We will continue with these two models in order to get quick results

### Most Common Words Removal

```python
#finding the most common words in both classes ("good" and "not good") separately:

freq3=[]
freq1 = pd.Series(' '.join(df_train['clean_text'][df_train['rating_class']==1]).split()).value_counts()[:80]
freq1=list(freq1.index)
freq2 = pd.Series(' '.join(df_train['clean_text'][df_train['rating_class']==0]).split()).value_counts()[:80]
freq2=list(freq2.index)

# finding matching words in both list and creating a new list for mathing words.

for w in freq2:
    if w in freq1:
        freq3.append(w)

# checking the mathing ration
print('Match-up ration {}: '.format(len(freq3)/len(freq1)))
```

```python
# removing the "no" and "not" words from the mathing words
most_common_words = freq3
most_common_words.remove('no')
most_common_words.remove('not')

# create a function in order to remove most common words from the corpus
def remove_most_comman(text, stop_words=most_common_words):
    tokens = [w for w in nltk.word_tokenize(text) if w not in stop_words]
    return ' '.join(tokens)

# applying the "remove_most_common" function to the "clean_text" and create a new feature.
df_train['clean_text_mc'] = df_train['clean_text'].map(lambda text: remove_most_comman(text))
```

```python
# checking the extension of the removal of non-English words
from nltk.tokenize import word_tokenize
tokens1=len([w for t in (df_train["clean_text"].apply(word_tokenize)) for w in t])
tokens2=len([w for t in (df_train["clean_text_mc"].apply(word_tokenize)) for w in t])
print('Number of clean tokens: {}\n'.format(tokens2))
print('Percentage of removed tokens: {0:.2f}'.format(1-(tokens2/tokens1)))
```

```python
X_2 = df_train['clean_text_mc']
y_2 = df_train['rating_class']
X_train2, X_test2, y_train2, y_test2 = train_test_split(X_2, y_2, test_size=0.25, random_state=5)

cv2 = CountVectorizer(ngram_range=(1,1))
cv_train2 = cv2.fit_transform(X_train2)
cv_train2 = cv_train2.toarray()
cv_test2 = cv2.transform(X_test2)
cv_test2 = cv_test2.toarray()
```

```python
clf1 = LogisticRegression(random_state=1)
clf2 = SVC(kernel = 'linear')

labels = ['Logistic Regression', 'Linear SVM']
print('\nMost Common Words Removal\n\n')

for clf, label in zip([clf1, clf2], labels):
    clf.fit(cv_train2, y_train2)
    y_pred_clf = clf.predict(cv_test2)
    cm = confusion_matrix(y_test2, y_pred_clf)
    
    print('[{}]\n'.format(label))
    print('1. Accuarcy: {}\n'.format(metrics.accuracy_score(y_test2, y_pred_clf)))
    print('2. The F-1 score of the model {}\n'.format(f1_score(y_test2, y_pred_clf, average='weighted')))
    print('3. The recall score of the model {}\n'.format(recall_score(y_test2, y_pred_clf, average='weighted')))
    print('4. Classification Report:\n{}\n5. Confusion matrix:\n{}'.format
          (classification_report(y_test2, y_pred_clf), cm))
```

#### Result of most common words removal:
    * Contribution to best f-1 score               : 0.3% (almost nothing)
    
    * Contribution to best f-1 score of minor class: 0 %

### Removal of Words with one letter and more than 15 letters 

```python
def remove_ones_and_longs(text):
    text = re.sub(r'\b\w{1,2}\b', '', text)
    text = re.sub(r'\b\w{15,}\b', '', text)
    text = re.sub(' +', ' ', text)
    text = text.strip()
    return text

df_train['clean_text_ones'] = df_train['clean_text'].map(lambda text: remove_ones_and_longs(text))
```

```python
# Let's put aside number of raw tokens in order to measure of cleaned tokens
from nltk.tokenize import word_tokenize
tokens1=len([w for t in (df_train["clean_text"].apply(word_tokenize)) for w in t])
tokens2=len([w for t in (df_train["clean_text_ones"].apply(word_tokenize)) for w in t])
print('Number of clean tokens: {}\n'.format(tokens2))
print('Percentage of removed tokens: {0:.2f}'.format(1-(tokens2/tokens1)))
```

```python
X_4 = df_train['clean_text_ones']
y_4 = df_train['rating_class']
X_train4, X_test4, y_train4, y_test4 = train_test_split(X_4, y_4, test_size=0.25, random_state=5)

cv4 = CountVectorizer(ngram_range=(1,1))
cv_train4 = cv4.fit_transform(X_train4)
cv_train4 = cv_train4.toarray()
cv_test4 = cv4.transform(X_test4)
cv_test4 = cv_test4.toarray()
```

```python
clf1 = LogisticRegression(random_state=1)
clf2 = SVC(kernel = 'linear')

labels = ['Logistic Regression', 'Linear SVM']
print('Short and Long Words Removal\n\n')

for clf, label in zip([clf1, clf2], labels):
    clf.fit(cv_train4, y_train4)
    y_pred_clf = clf.predict(cv_test4)
    cm = confusion_matrix(y_test4, y_pred_clf)
    
    print('********** [{}] **********\n'.format(label))
    print('1. Accuarcy: {}\n'.format(metrics.accuracy_score(y_test4, y_pred_clf)))
    print('2. The F-1 score of the model {}\n'.format(f1_score(y_test4, y_pred_clf, average='weighted')))
    print('3. The recall score of the model {}\n'.format(recall_score(y_test4, y_pred_clf, average='weighted')))
    print('4. Classification Report:\n{}\n5. Confusion matrix:\n{}\n'.format
          (classification_report(y_test4, y_pred_clf), cm))
```
